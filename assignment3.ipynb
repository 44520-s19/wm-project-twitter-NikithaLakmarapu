```python
import tweepy
from twitter_keys import *
import pickle
auth=tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api=tweepy.API(auth)
donald_tweets=api.user_timeline('realDonaldTrump')
with open('trump.pkl','wb') as f:
    pickle.dump(donald_tweets, f)
barack_tweets=api.user_timeline('BarackObama')
with open('obama.pkl','wb') as f:
    pickle.dump(barack_tweets, f)
hillary_tweets=api.user_timeline('HillaryClinton')
with open('clinton.pkl','wb') as f:
    pickle.dump(hillary_tweets, f)
```

### Analysis steps
##### Attributes considered:
1. Retweet value for each tweet
1. Word_count for each tweet

 For each twitter account, appending retweet values of all tweets to an array .
 
 For each twitter account, appending all the values of word_count(each tweet) of all tweets into another array.
 
 Printing the two array lists for each twitter account.


```python
#retweet_count && no of words for each tweet
donald_retweet=[]
donald_twords=[]
for i in range(len(donald_tweets)):
    donald_retweet.append(donald_tweets[i].retweet_count)
    donald_twords.append(len(donald_tweets[i].text.split()))    
print(donald_retweet)
print(donald_twords)

barack_retweet=[]
barack_twords=[]
for i in range(len(barack_tweets)):
    barack_retweet.append(barack_tweets[i].retweet_count)
    barack_twords.append(len(barack_tweets[i].text.split()))   
print(list(barack_retweet))
print(barack_twords)

hillary_retweet=[]
hillary_twords=[]
for i in range(len(hillary_tweets)):
    hillary_retweet.append(hillary_tweets[i].retweet_count)
    hillary_twords.append(len(hillary_tweets[i].text.split()))   
print(list(hillary_retweet))
print(hillary_twords)
```

```python

import matplotlib.pyplot as plt
xd = donald_retweet
yd =donald_twords
xb=barack_retweet
yb=barack_twords
xh=hillary_retweet
yh=hillary_twords
plt.scatter(xd, yd,color='r',marker='1',label='Trump')
plt.scatter(xb, yb,color='b',marker='P',label='Obama')
plt.scatter(xh,yh,color='g',marker='8',label='Clinton')
plt.xlabel('Retweets')
plt.ylabel('Words_per_tweet')
plt.title('Retweets vs Words per tweet')
plt.legend(loc='lower right',shadow=True)
plt.show()
```
